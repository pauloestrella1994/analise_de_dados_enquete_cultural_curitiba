```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import collections
```

```python
dados_area_de_atuacao = pd.read_csv('Dados_gerais.csv')
```

```python
#Area de Atuação

area_de_atuacao = dados_area_de_atuacao["Qual a área de atuação do espaço?"]
lista = []
for palavra in area_de_atuacao:
    lista.append(palavra)

area_de_atuacao_string = ','.join(lista)

danca = area_de_atuacao_string.count('Dança')
artes_visuais = area_de_atuacao_string.count('Artes Visuais')
literatura = area_de_atuacao_string.count('Literatura')
musica = area_de_atuacao_string.count('Música')
teatro = area_de_atuacao_string.count('Teatro')
patrimonio_cultural = area_de_atuacao_string.count('Patrimônio Cultural')
audiovisual = area_de_atuacao_string.count('Audiovisual')
artesanato = area_de_atuacao_string.count('Artesanato')
circo = area_de_atuacao_string.count('Circo')
outros = area_de_atuacao_string.count('musculação') + area_de_atuacao_string.count('Editora') + area_de_atuacao_string.count('Produtora Cultural') + area_de_atuacao_string.count('Novas Mídias') + area_de_atuacao_string.count('tecnologias open source') + area_de_atuacao_string.count('arte-tecnologia') + area_de_atuacao_string.count('arte-tecnologia')


data_areas = {
    'Outros': outros,     
    'Dança': danca, 
    'Artes Visuais': artes_visuais,
    'Literatura': literatura,
    'Música': musica,
    'Teatro': teatro,
    'Patrimônio Cultural': patrimonio_cultural,
    'AudioVisual': audiovisual,
    'Artesanato': artesanato,
    'Circo': circo,   
}
group_data = list(data_areas.values())
group_names = list(data_areas.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots()
ax.barh(group_names, group_data)
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (8,5)
plt.title('Atividades desenvolvidas')
```

```python
#Quantidade de Pessoas que trabalham frequentemente
dados_receita = pd.read_csv('Dados_de_receita.csv')
dados_receita.hist(column='Quantas pessoas trabalham direta e periodicamente no Espaço?', bins=50)
plt.xlabel('Número de trabalhadores')
plt.ylabel('Frequência de espaços')
plt.show()
```

```python
#Frequencia de custos de aluguel
dados_aluguel = [i for i in dados_receita['Em caso locação, qual o custo médio do aluguel?']]
freq_aluguel = collections.Counter(dados_aluguel)
x_aluguel = list(freq_aluguel.keys())
label_aluguel = [i + ' reais' if i != 'Não informado' else i for i in x_aluguel]

y_aluguel = list(freq_aluguel.values())

fig1, ax1 = plt.subplots()
ax1.pie(y_aluguel, labels=label_aluguel,autopct='%1.1f%%',shadow=True, startangle=100)
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title('Gastos com aluguel')
```

```python
#Frequencia de manutenção
dados_manutencao = [i for i in dados_receita['Qual a média do custo mensal de manutenção da sede? (água, luz, telefone e internet)']]
freq_manutencao = collections.Counter(dados_manutencao)
x_manutencao = list(freq_manutencao.keys())
label_manutencao = [i + ' reais' for i in x_manutencao]

y_manutencao = list(freq_manutencao.values())

fig1, ax1 = plt.subplots()
ax1.pie(y_manutencao, labels=label_manutencao,autopct='%1.1f%%',shadow=True, startangle=100)
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title('Gastos com manutenções mensais')
```

```python
#Gestao de receita
dados_gestao = dados_receita["Quais as formas de gestão de receitas do espaço?"]

dados_gestao_lista = [i for i in dados_gestao]
dados_gestao_receitas = ','.join(dados_gestao_lista)

bilheteria = dados_gestao_receitas.count('Bilheteria')
cursos = dados_gestao_receitas.count('Cursos')
aluguel_equipamento_espaco = dados_gestao_receitas.count('Aluguel de equipamentos ou espaço')
bar_cafe = dados_gestao_receitas.count('Bar/café')
doacao = dados_gestao_receitas.count('Doação')
editais = dados_gestao_receitas.count('Editais')
outros = dados_gestao_receitas.count('venda de livros em oficinas') + dados_gestao_receitas.count('Aulas') + dados_gestao_receitas.count('servicos') + dados_gestao_receitas.count('Mensalidade') + dados_gestao_receitas.count('Vendas de apresentações teatrais') + dados_gestao_receitas.count('Recursos provenientes de apresentações em escolas e empresas') + dados_gestao_receitas.count('oficina de bicicletas') + dados_gestao_receitas.count('malojinha') + dados_gestao_receitas.count('ma mala a qual levamos artesanatos fabricados por nós e vendemos para fazer renda complementar')


data_gestao_receitas = {
    'Outros': outros,     
    'Bilheteria': bilheteria, 
    'Cursos': cursos,
    'Aluguel de equipamentos ou espaço': aluguel_equipamento_espaco,
    'Bar/Café': bar_cafe,
    'Doação': doacao,
    'Editais': editais,  
}

group_data_gestao = list(data_gestao_receitas.values())
group_names_gestao = list(data_gestao_receitas.keys())
group_mean = np.mean(group_data_gestao)

fig, ax = plt.subplots()
ax.barh(group_names_gestao, group_data_gestao)
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (8,5)
plt.title('Gestão de receitas')
```

```python
#Parcerias
dados_relativos = pd.read_csv('Dados_relativos.csv')
dados_rl_parceria = dados_relativos["O espaço já estabeleceu algum tipo de parceria?"]

df_list = [i for i in dados_rl_parceria]
df_parceria = ','.join(df_list)

sem_parcerias = df_parceria.count('Não')
escola = df_parceria.count('escolas')
bibliotecas = df_parceria.count('bibliotecas')
associacao = df_parceria.count('associações de moradores')
adm_publica = df_parceria.count('administração pública')
empresas = df_parceria.count('empresa privada')

dados_parceria = {
    'Sem parcerias': sem_parcerias,
    'Escolas': escola,
    'Bibliotecas': bibliotecas,
    'Administração pública': adm_publica,
    'Empresas Privadas': empresas
}

group_data_parcerias = list(dados_parceria.values())
group_names_parcerias = list(dados_parceria.keys())

fig1, ax1 = plt.subplots()
ax1.pie(group_data_parcerias, labels=group_names_parcerias,autopct='%1.1f%%',shadow=True, startangle=100)
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title('Parcerias')
```

```python
#Dificuldades antes da pandemia
dados_antes_pandemia = dados_relativos['Quais as maiores dificuldades enfrentadas pelo seu espaço antes da pandemia?']

df_list_antes_pandemia = [i for i in dados_antes_pandemia]
df_antes_pandemia = ','.join(df_list_antes_pandemia)

aluguel = df_antes_pandemia.count('aluguel do espaço')
manutencao = df_antes_pandemia.count('manutenção do espaço físico')
salario = df_antes_pandemia.count('salário dos funcionários')
despesas_adm = df_antes_pandemia.count('despesas administrativas')
conteudo_virtual = df_antes_pandemia.count('Produção de conteúdo virtual')
redes_sociais = df_antes_pandemia.count('Alimentação de redes sociais')
formacao_publico = df_antes_pandemia.count('Formação de público')
politicas_publicas = df_antes_pandemia.count('Insuficiência de políticas públicas')
fidelizacao = df_antes_pandemia.count('Fidelização de novos alunos')


dados_dif_pre_pandemia = {
    'Aluguel': aluguel,
    'Manutenção das dependências': manutencao,
    'Salários': salario,
    'Despesas Administrativas': despesas_adm,
    'Produção de conteúdo': conteudo_virtual,
    'Manutenção de redes sociais': redes_sociais,
    'Formação de Público': formacao_publico,
    'Ineficiência de Políticas Públicas': politicas_publicas,
    'Fidelização de novos alunos': fidelizacao
}

group_data_dificuldades_pre = list(dados_dif_pre_pandemia.values())
group_names_dificuldades_pre = list(dados_dif_pre_pandemia.keys())

fig, ax = plt.subplots()
ax.barh(group_names_dificuldades_pre, group_data_dificuldades_pre)
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (10,10)
plt.title('Dificuldades antes da pandemia')
```

```python
#Dificuldades na pandemia
dados_depois_pandemia = dados_relativos['Quais as maiores dificuldades enfrentadas pelo seu espaço com a pandemia?']

df_list_pandemia = [i for i in dados_depois_pandemia]
df_pandemia = ','.join(df_list_pandemia)

aluguel_pan = df_pandemia.count('aluguel do espaço')
manutencao_pan = df_pandemia.count('manutenção do espaço físico')
salario_pan = df_pandemia.count('salário dos funcionários')
despesas_adm_pan = df_pandemia.count('despesas administrativas')
conteudo_virtual_pan = df_pandemia.count('Produção de conteúdo virtual')
redes_sociais_pan = df_pandemia.count('Alimentação de redes sociais')
formacao_publico_pan = df_pandemia.count('Formação de público')
politicas_publicas_pan = df_pandemia.count('Insuficiência de políticas públicas')
fidelizacao_pan = df_pandemia.count('Fidelização de novos alunos')
eventos_pan = df_pandemia.count('realização de eventos')
aulas_cursos_pan = df_pandemia.count('aulas e cursos')
ensaios_pan = df_pandemia.count('realização de ensaios')
normas_pan = df_pandemia.count('normas sanitárias')
reducao_alunos_pan = df_pandemia.count('Redução de alunos')
reducao_doacao_pan = df_pandemia.count('financeira de doações')
reducao_cafe_bar_pan = df_pandemia.count('financeira do café/bar')


dados_pandemia = {
    'Aluguel': aluguel_pan,
    'Manutenção das dependências': manutencao_pan,
    'Salários': salario_pan,
    'Despesas Administrativas': despesas_adm_pan,
    'Produção de conteúdo': conteudo_virtual_pan,
    'Manutenção de redes sociais': redes_sociais_pan,
    'Formação de Público': formacao_publico_pan,
    'Ineficiência de Políticas Públicas': politicas_publicas_pan,
    'Fidelização de novos alunos': fidelizacao_pan,
    'Fechamento para Eventos': eventos_pan,
    'Fechamento pra aulas/cursos': aulas_cursos_pan,
    'Fechamento pra Ensaios': ensaios_pan,
    'Adequação às Normas Sanitárias': normas_pan,
    'Redução de Alunos': reducao_alunos_pan,
    'Redução de Doações': reducao_doacao_pan,
    'Redução de Café/Bar': reducao_cafe_bar_pan
}

group_data_dificuldades_pan = list(dados_pandemia.values())
group_names_dificuldades_pan = list(dados_pandemia.keys())

fig, ax = plt.subplots()
ax.barh(group_names_dificuldades_pan, group_data_dificuldades_pan)
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (10,10)
plt.title('Dificuldades na pandemia')
```

```python
#word cloud
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

nltk.download('stopwords')

dados_relativos = pd.read_csv('Dados_relativos.csv')
dados_agravamento = dados_relativos['No caso de agravamento das dificuldades pelo seu espaço, descreva quais foram identificadas. '].dropna()
stop_words = set(stopwords.words("portuguese"))
other_words = ['etc.', '.', ',', ';', '\n', 'de', 'em', 'por', 'um', 'como', 'e', 'a', 'ou', 'dos', 'o', 'na', 'os', 'dois']
stop_words.update(other_words)

clean_data = [i.lower() for i in dados_agravamento]

data = []

for i in clean_data:
    if i not in stop_words:
        data.append(i)
        
fdist = FreqDist(data)
fdist.most_common(20)


wc = WordCloud()
img = wc.generate_from_text(' '.join(data))
plt.show(img)
```
