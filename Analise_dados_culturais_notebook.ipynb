```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import collections
```

```python
dados_area_de_atuacao = pd.read_csv('Dados_gerais.csv')
```

```python
#Area de Atuação

area_de_atuacao = dados_area_de_atuacao["Qual a área de atuação do espaço?"]
lista = []
for palavra in area_de_atuacao:
    lista.append(palavra)

area_de_atuacao_string = ','.join(lista)

danca = area_de_atuacao_string.count('Dança')
artes_visuais = area_de_atuacao_string.count('Artes Visuais')
literatura = area_de_atuacao_string.count('Literatura')
musica = area_de_atuacao_string.count('Música')
teatro = area_de_atuacao_string.count('Teatro')
patrimonio_cultural = area_de_atuacao_string.count('Patrimônio Cultural')
audiovisual = area_de_atuacao_string.count('Audiovisual')
artesanato = area_de_atuacao_string.count('Artesanato')
circo = area_de_atuacao_string.count('Circo')
outros = area_de_atuacao_string.count('musculação') + area_de_atuacao_string.count('Editora') + area_de_atuacao_string.count('Produtora Cultural') + area_de_atuacao_string.count('Novas Mídias') + area_de_atuacao_string.count('tecnologias open source') + area_de_atuacao_string.count('arte-tecnologia') + area_de_atuacao_string.count('arte-tecnologia')


data_areas = {
    'Outros': outros,     
    'Dança': danca, 
    'Artes Visuais': artes_visuais,
    'Literatura': literatura,
    'Música': musica,
    'Teatro': teatro,
    'Patrimônio Cultural': patrimonio_cultural,
    'AudioVisual': audiovisual,
    'Artesanato': artesanato,
    'Circo': circo,   
}
group_data = list(data_areas.values())
group_names = list(data_areas.keys())
group_mean = np.mean(group_data)

fig, ax = plt.subplots()
ax.barh(group_names, group_data)
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (8,5)
plt.title('Atividades desenvolvidas')
```

```python
#Quantidade de Pessoas que trabalham frequentemente
dados_receita = pd.read_csv('Dados_de_receita.csv')
dados_receita.hist(column='Quantas pessoas trabalham direta e periodicamente no Espaço?', bins=50)
plt.xlabel('Número de trabalhadores')
plt.ylabel('Frequência de espaços')
plt.show()
```

```python
#Frequencia de custos de aluguel
dados_aluguel = [i for i in dados_receita['Em caso locação, qual o custo médio do aluguel?']]
freq_aluguel = collections.Counter(dados_aluguel)
x_aluguel = list(freq_aluguel.keys())
label_aluguel = [i + ' reais' if i != 'Não informado' else i for i in x_aluguel]

y_aluguel = list(freq_aluguel.values())

fig1, ax1 = plt.subplots()
ax1.pie(y_aluguel, labels=label_aluguel,autopct='%1.1f%%',shadow=True, startangle=100)
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title('Gastos com aluguel')
```

```python
#Frequencia de manutenção
dados_manutencao = [i for i in dados_receita['Qual a média do custo mensal de manutenção da sede? (água, luz, telefone e internet)']]
freq_manutencao = collections.Counter(dados_manutencao)
x_manutencao = list(freq_manutencao.keys())
label_manutencao = [i + ' reais' for i in x_manutencao]

y_manutencao = list(freq_manutencao.values())

fig1, ax1 = plt.subplots()
ax1.pie(y_manutencao, labels=label_manutencao,autopct='%1.1f%%',shadow=True, startangle=100)
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.title('Gastos com manutenções mensais')
```

```python
#Gestao de receita
dados_gestao = dados_receita["Quais as formas de gestão de receitas do espaço?"]

dados_gestao_lista = [i for i in dados_gestao]
dados_gestao_receitas = ','.join(dados_gestao_lista)

bilheteria = dados_gestao_receitas.count('Bilheteria')
cursos = dados_gestao_receitas.count('Cursos')
aluguel_equipamento_espaco = dados_gestao_receitas.count('Aluguel de equipamentos ou espaço')
bar_cafe = dados_gestao_receitas.count('Bar/café')
doacao = dados_gestao_receitas.count('Doação')
editais = dados_gestao_receitas.count('Editais')
outros = dados_gestao_receitas.count('venda de livros em oficinas') + dados_gestao_receitas.count('Aulas') + dados_gestao_receitas.count('servicos') + dados_gestao_receitas.count('Mensalidade') + dados_gestao_receitas.count('Vendas de apresentações teatrais') + dados_gestao_receitas.count('Recursos provenientes de apresentações em escolas e empresas') + dados_gestao_receitas.count('oficina de bicicletas') + dados_gestao_receitas.count('malojinha') + dados_gestao_receitas.count('ma mala a qual levamos artesanatos fabricados por nós e vendemos para fazer renda complementar')


data_gestao_receitas = {
    'Outros': outros,     
    'Bilheteria': bilheteria, 
    'Cursos': cursos,
    'Aluguel de equipamentos ou espaço': aluguel_equipamento_espaco,
    'Bar/Café': bar_cafe,
    'Doação': doacao,
    'Editais': editais,  
}

group_data_gestao = list(data_gestao_receitas.values())
group_names_gestao = list(data_gestao_receitas.keys())
group_mean = np.mean(group_data_gestao)

fig, ax = plt.subplots()
ax.barh(group_names_gestao, group_data_gestao)
plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (8,5)
plt.title('Gestão de receitas')
```
